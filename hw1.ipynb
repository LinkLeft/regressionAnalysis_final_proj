## 第一题

### 第一问

- E[$\epsilon_i$] = 0
- Var($\epsilon_i$) = $\sigma^2$
- Cov($\epsilon_i, \epsilon_j$) = 0

### 第二问

$\epsilon$包含了因变量中未被自变量描述的部分，是其它未被自变量包含的影响因素的总和

### 第三问

$\beta_0$和$\beta_1$是线性回归模型中截距和斜率的真实值，而$\hat{\beta}_0$和$\hat{\beta}_1$是基于数据得到的拟合值。
残差$e_i$是根据拟合值得来的，是真实值与拟合值的差，而误差$\epsilon_i$是根据真实值得来的，由于真实值未知，其具体值无法求出

### 第四问

进行假设检验能够验证模型假设的正确性，判断因变量与自变量间是否有显著的回归关系

## 第二题

$$\hat{\beta}_0=\bar{y}-\hat{\beta}_1\bar{x}=\bar{y}-\bar{x}\sum_{i=1}^n{\frac{x_i-\bar{x}}{\sum_{j=1}^n{\text{（}x_j-\bar{x}\text{）}^2}}y_i}\,\,
\\
=\,\,\sum_{i=1}^n{\frac{y_i}{n}\,\,-\,\,\bar{x}\sum_{i=1}^n{\frac{x_i-\bar{x}}{\sum_{j=1}^n{\text{（}x_j-\bar{x}\text{）}^2}}y_i}}
\\
=\,\,\sum_{i=1}^n{\left( \frac{1}{n}\,\,-\,\,\frac{\bar{x}\left( x_i-\bar{x} \right)}{\sum_{j=1}^n{\left( x_j-\bar{x} \right) ^2}} \right) y_i}$$

## 第三题

$$SST=\sum_{i=1}^n{\text{（}y_i-\bar{y}\text{）}^2}
\\
\phantom{SST}=\sum_{i=1}^n{\left( y_i-\hat{y}+\hat{y}-\bar{y} \right) ^2}
\\
\phantom{SST}=\sum_{i=1}^n{\left( y_i-\hat{y} \right) ^2+\sum_{i=1}^n{\left( \hat{y}-\bar{y} \right) ^2+2\sum_{i=1}^n{\left( y_i-\hat{y} \right) \left( \hat{y}-\bar{y} \right)}}}
\\
\phantom{SST}=SSR+SSE+2\sum_{i=1}^n{\left( y_i-\hat{y} \right) \left( \hat{y}-\bar{y} \right)}$$

$$\sum_{i=1}^n{\text{（}y_i-\hat{y}\text{）}\left( \hat{y}-\bar{y} \right) =\sum_{i=1}^n{e_i\left( \hat{\beta}_0+\hat{\beta}_1x_i-\bar{y} \right)}}
\\
=\left( \hat{\beta}_0-\bar{y} \right) \sum_{i=1}^n{e_i+\hat{\beta}_1\sum_{i=1}^n{e_ix_i}}
\\
=0$$

得$$SST=SSR+SSE$$

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.uniform ( low = 0 , high = 10 , size = 1000) # input X
epsilon = np.random.normal(loc = 0, scale = 1, size = 1000)
y = 2 * x + 3 + epsilon
```

```python
T = 100
n = 10
beta1_arr = []
beta0_arr = []

for t in range(T):
    sample_idx = np.random.choice(1000, size = n, replace=False)
    x_sample = x[sample_idx]
    y_sample = y[sample_idx]

    x_mean = np.mean(x_sample)
    y_mean = np.mean(y_sample)

    fenzi = np.sum((x_sample - x_mean) * (y_sample - y_mean))
    fenmu = np.sum((x_sample - x_mean) ** 2)

    beta1 = fenzi / fenmu
    beta0 = y_mean - beta1 * x_mean

    beta1_arr.append(beta1)
    beta0_arr.append(beta0)

arr1 = np.array(beta1_arr)
arr0 = np.array(beta0_arr)

beta1_mean = np.mean(arr1)
beta0_mean = np.mean(arr0)
beta1_var = np.var(arr1)
beta0_var = np.var(arr0)

# Unicode 编码
beta = "\u03B2"  # β
hat = "\u0302"   # ̂
sub_0 = "\u2080" # ₀
sub_1 = "\u2081" # ₁

# 动态生成 β̂₀ 和 β̂₁
beta_0_hat = f"{beta}{hat}{sub_0}"
beta_1_hat = f"{beta}{hat}{sub_1}"

print(f"当n = {n}")
print(f"{beta_1_hat}的均值为{beta1_mean},方差为{beta1_var}")
print(f"{beta_0_hat}的均值为{beta0_mean},方差为{beta0_var}")
print("作出散点图如下：")

plt.scatter(arr0, arr1)
plt.title(f"n = {n}")
plt.xlabel(f"{beta_0_hat}")
plt.ylabel(f"{beta_1_hat}")

plt.show()


    


```

## 第五题
### 第一问

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from scipy.stats import f

x_arr = np.array([1000, 820, 200, 750, 630, 480, 900, 320, 250, 800])
y_arr = np.array([5.0, 4.3, 1.2, 4.0, 3.5, 2.7, 4.8, 2.0, 1.5, 4.2])

plt.scatter(x_arr, y_arr)
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

从散点图中可看出x与y间大致呈线性关系

### 第二问

```python
x_mean = np.mean(x_arr)
y_mean = np.mean(y_arr)

fenzi = np.sum((x_arr - x_mean) * (y_arr - y_mean))
fenmu = np.sum((x_arr - x_mean) ** 2)

beta1 = fenzi / fenmu
beta0 = y_mean - beta1 * x_mean

print(f"回归方程为: y = {beta1}x + {beta0}")
```

### 第三问

```python
y_pred = np.array(beta1 * x_arr + beta0)
residuals = y_arr - y_pred
sum_of_residuals = np.sum(residuals ** 2)

n = len(y_arr)
reg_std_error = np.sqrt(sum_of_residuals / (n - 2))

print(f"回归标准误差为{reg_std_error}")
```

### 第四问

```python
# 计算回归系数的标准误
SE_beta1 = reg_std_error / np.sqrt(fenmu)
SE_beta0 = reg_std_error * np.sqrt(1/n + x_mean**2/fenmu)

# 计算t分布的临界值（95%置信水平）
alpha = 0.05
df = n - 2
t_critical = t.ppf(1 - alpha/2, df)

# 计算置信区间
beta1_conf_int = (beta1 - t_critical * SE_beta1, beta1 + t_critical * SE_beta1)
beta0_conf_int = (beta0 - t_critical * SE_beta0, beta0 + t_critical * SE_beta0)

print("斜率beta1的95%置信区间为:", beta1_conf_int)
print("截距beta0的95%置信区间为:", beta0_conf_int)
```

### 第五问

```python
SSR = np.sum((y_pred - y_mean) ** 2)
SSE = np.sum((y_arr - y_pred) ** 2)

F = SSR / (SSE / (n - 2))

df1 = 1  # 自变量数量
df2 = n - 2  # 残差自由度
p_value = 1 - f.cdf(F, df1, df2)

print("F统计量:", F)
print("p值:", p_value)
print("在5%显著性水平下拒绝零假设，认为因变量y与自变量x间存在线性相关关系")
```

### 第六问

H0：$\beta_1 = 0$ H1:$\beta_1 \neq 0$

由第二问与第四问的结果,$\beta_1 = 0$的值落于95\%置信区间中，知在95\%的置信水平下接受原假设，认为$\beta_1 = 0$
